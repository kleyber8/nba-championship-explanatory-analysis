 **Analisis de la veracidad del datasets**


```python
import pandas as pd

df = pd.read_csv(r"C:\Users\Max\Documents\GitHub\nba_playoffs_predictor\data\raw\all_playoffs_2020_2025.csv")
# Ver columnas
print("Columnas:", df.columns.tolist())
print("Primeras filas:")
print(df.head())

# Tomar un game_id cualquiera para inspeccionar
gid = df['GAME_ID'].iloc[0]
game = df[df['GAME_ID'] == gid]
print("\nPartido ejemplo:")
print(game[['GAME_ID', 'TEAM_NAME', 'PTS']])

# Calcular diferencia
diff = game['PTS'].diff().abs().iloc[-1]
print(f"Diferencia de puntos: {diff}")
```

```python
# Contar partidos con diferencia absoluta de puntos
differences = []
for gid, group in df.groupby('GAME_ID'):
    if len(group) == 2:
        pts1 = group['PTS'].iloc[0]
        pts2 = group['PTS'].iloc[1]
        diff = abs(pts1 - pts2)
        differences.append(diff)
    else:
        print(f"GAME_ID {gid} tiene {len(group)} filas (raro)")

print(f"Total de partidos únicos: {len(differences)}")
print(f"Partidos con diff ≤5: {sum(1 for d in differences if d <= 5)}")
print(f"Partidos con diff ≤10: {sum(1 for d in differences if d <= 10)}")
print(f"Partidos con diff >10: {sum(1 for d in differences if d > 10)}")
print(f"Diferencias: min={min(differences)}, max={max(differences)}")
```
